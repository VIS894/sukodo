```python
# Imports
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import mplfinance as mpf
import unittest
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import pytz

# Global Variables
ticker_names = {
    'GOOGL': 'Alphabet Inc. (Class A)', 'TSLA': 'Tesla Inc.',
    'AAPL': 'Apple Inc.', 'AMZN': 'Amazon.com Inc.',
    'META': 'Meta Platforms Inc.', 'MSFT': 'Microsoft Corporation',
    'NVDA': 'NVIDIA Corporation', 'AMD': 'Advanced Micro Devices',

    # Other Notable Stocks
    'BABA': 'Alibaba Group', 'CORZ': 'Core Scientific',
    'CRWD': 'CrowdStrike Holdings', 'F': 'Ford Motor Company',
    'GM': 'General Motors', 'RACE': 'Ferrari NV',
    'HOOD': 'Robinhood Markets', 'LOVE': 'The Lovesac Company',
    'RKLB': 'Rocket Lab USA', 'LULU': 'Lululemon Athletica'
}
failed_tickers = []

# Technical Indicator Functions
def check_technical_indicators(ticker):
    try:
        end_date = datetime.now(pytz.UTC)
        start_date = end_date - timedelta(days=250)
        data = yf.download(ticker, start=start_date.strftime('%Y-%m-%d'), end=end_date.strftime('%Y-%m-%d'))
        if data.empty:
            failed_tickers.append(ticker)
            return None

        # Compute moving averages
        data['200EMA'] = data['Close'].ewm(span=200, adjust=False).mean()
        data['20MA'] = data['Close'].rolling(window=20).mean()
        data['50MA'] = data['Close'].rolling(window=50).mean()

        # RSI Calculation
        delta = data['Close'].diff()
        gain = delta.where(delta > 0, 0).rolling(window=14).mean()
        loss = -delta.where(delta < 0, 0).rolling(window=14).mean()
        rs = gain / loss
        data['RSI'] = 100 - (100 / (1 + rs))

        # MACD Calculation
        data['MACD'] = data['Close'].ewm(span=12, adjust=False).mean() - data['Close'].ewm(span=26, adjust=False).mean()
        data['Signal_Line'] = data['MACD'].ewm(span=9, adjust=False).mean()
        data['MACD_Histogram'] = data['MACD'] - data['Signal_Line']

        # Stochastic Oscillator
        low_14 = data['Low'].rolling(window=14).min()
        high_14 = data['High'].rolling(window=14).max()
        data['Stochastic_K'] = (data['Close'] - low_14) / (high_14 - low_14) * 100
        data['Stochastic_D'] = data['Stochastic_K'].rolling(window=3).mean()

        # OBV Calculation
        data['OBV'] = (data['Volume'] * ((data['Close'].diff() > 0) * 2 - 1)).cumsum()
        data['Avg_Volume'] = data['Volume'].rolling(window=10).mean()

        # Bollinger Bands
        data['BB_Middle'] = data['Close'].rolling(window=20).mean()
        data['BB_Std'] = data['Close'].rolling(window=20).std()
        data['BB_Upper'] = data['BB_Middle'] + 2 * data['BB_Std']
        data['BB_Lower'] = data['BB_Middle'] - 2 * data['BB_Std']

        # Average True Range (ATR)
        high_low = data['High'] - data['Low']
        high_close = np.abs(data['High'] - data['Close'].shift())
        low_close = np.abs(data['Low'] - data['Close'].shift())
        tr = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)
        data['ATR'] = tr.rolling(window=14).mean()

        # Relative Volatility Index (RVI)
        std = data['Close'].rolling(window=10).std()
        data['RVI'] = (data['Close'].diff() / std).ewm(span=10, adjust=False).mean() * 100

        # ADX (Average Directional Index)
        data['Plus_DM'] = np.where((data['High'] - data['High'].shift(1)) > (data['Low'].shift(1) - data['Low']),
                                   (data['High'] - data['High'].shift(1)), 0)
        data['Minus_DM'] = np.where((data['Low'].shift(1) - data['Low']) > (data['High'] - data['High'].shift(1)),
                                    (data['Low'].shift(1) - data['Low']), 0)
        tr_smooth = tr.rolling(window=14).mean()
        data['Plus_DI'] = 100 * (data['Plus_DM'].rolling(window=14).mean() / tr_smooth)
        data['Minus_DI'] = 100 * (data['Minus_DM'].rolling(window=14).mean() / tr_smooth)
        dx = 100 * np.abs(data['Plus_DI'] - data['Minus_DI']) / (data['Plus_DI'] + data['Minus_DI'])
        data['ADX'] = dx.ewm(span=14, adjust=False).mean()

        # Extract latest values
        latest_close = data['Close'].iloc[-1].item()
        latest_200ema = data['200EMA'].iloc[-1].item()
        latest_20ma = data['20MA'].iloc[-1].item()
        latest_50ma = data['50MA'].iloc[-1].item()
        latest_rsi = data['RSI'].iloc[-1].item()
        latest_macd_hist = data['MACD_Histogram'].iloc[-1].item()
        prev_macd_hist = data['MACD_Histogram'].iloc[-2].item()
        latest_obv = data['OBV'].iloc[-1].item()
        prev_obv = data['OBV'].iloc[-2].item()
        latest_bb_upper = data['BB_Upper'].iloc[-1].item()
        latest_bb_lower = data['BB_Lower'].iloc[-1].item()
        latest_atr = data['ATR'].iloc[-1].item()
        latest_rvi = data['RVI'].iloc[-1].item()
        latest_adx = data['ADX'].iloc[-1].item()

        # Calculate percentage differences
        pct_diff_200ema = ((latest_close - latest_200ema) / latest_200ema) * 100
        pct_diff_20ma = ((latest_close - latest_20ma) / latest_20ma) * 100
        pct_diff_50ma = ((latest_close - latest_50ma) / latest_50ma) * 100

        # Define conditions
        volume_spike = data['Volume'].iloc[-1].item() > 1.5 * data['Avg_Volume'].iloc[-1].item()
        price_above_20ma = latest_close > latest_20ma
        macd_bullish = prev_macd_hist < 0 and latest_macd_hist > 0
        stochastic_bullish = data['Stochastic_K'].iloc[-2] < data['Stochastic_D'].iloc[-2] and \
                             data['Stochastic_K'].iloc[-1] > data['Stochastic_D'].iloc[-1] and \
                             data['Stochastic_K'].iloc[-1] < 30
        rsi_rebound = data['RSI'].iloc[-2].item() < 30 and latest_rsi > 30
        obv_rising = latest_obv > prev_obv
        bb_breakout = latest_close > latest_bb_upper or latest_close < latest_bb_lower
        strong_trend = latest_adx > 25

        bullish_reversal = (pct_diff_200ema < 0 and macd_bullish and stochastic_bullish and
                            price_above_20ma and (obv_rising or volume_spike))

        # Determine stock state
        if bullish_reversal:
            state = "Yes - Confirmed Bullish Reversal"
        elif (macd_bullish and stochastic_bullish and rsi_rebound):
            state = "Warning - Very Strong Reversal Building"
        elif (macd_bullish and stochastic_bullish) or (macd_bullish and rsi_rebound) or (stochastic_bullish and rsi_rebound):
            state = "Warning - Strong Reversal Building"
        elif macd_bullish or stochastic_bullish or rsi_rebound or bb_breakout or strong_trend:
            state = "Warning - Reversal Building"
        else:
            state = "No - Still Bearish"

        return {
            'Ticker': ticker,
            'Index Name': ticker_names.get(ticker, 'Unknown'),
            'Close': latest_close,
            '200EMA': latest_200ema,
            '20MA': latest_20ma,
            '50MA': latest_50ma,
            'Pct_Diff_200EMA': pct_diff_200ema,
            'Pct_Diff_20MA': pct_diff_20ma,
            'Pct_Diff_50MA': pct_diff_50ma,
            'RSI': latest_rsi,
            'OBV': latest_obv,
            'Volume Spike': volume_spike,
            'MACD Bullish': macd_bullish,
            'Stochastic Bullish': stochastic_bullish,
            'RSI Rebound': rsi_rebound,
            'OBV Rising': obv_rising,
            'BB_Upper': latest_bb_upper,
            'BB_Lower': latest_bb_lower,
            'ATR': latest_atr,
            'RVI': latest_rvi,
            'ADX': latest_adx,
            'BB Breakout': bb_breakout,
            'Strong Trend': strong_trend,
            'Data': data,
            'Bullish Reversal': bullish_reversal,
            'Stock State': state
        }
    except Exception as e:
        failed_tickers.append(ticker)
        return None

def backtest_technical_indicators(ticker):
    try:
        end_date = datetime.now(pytz.UTC)
        start_date = end_date - timedelta(days=730)  # 2 years of data
        data = yf.download(ticker, start=start_date.strftime('%Y-%m-%d'), end=end_date.strftime('%Y-%m-%d'))
        if data.empty:
            return None

        # Compute indicators
        data['200EMA'] = data['Close'].ewm(span=200, adjust=False).mean()
        data['20MA'] = data['Close'].rolling(window=20).mean()
        data['MACD'] = data['Close'].ewm(span=12, adjust=False).mean() - data['Close'].ewm(span=26, adjust=False).mean()
        data['Signal_Line'] = data['MACD'].ewm(span=9, adjust=False).mean()
        data['MACD_Histogram'] = data['MACD'] - data['Signal_Line']
        low_14 = data['Low'].rolling(window=14).min()
        high_14 = data['High'].rolling(window=14).max()
        data['Stochastic_K'] = (data['Close'] - low_14) / (high_14 - low_14) * 100
        data['Stochastic_D'] = data['Stochastic_K'].rolling(window=3).mean()
        delta = data['Close'].diff()
        gain = delta.where(delta > 0, 0).rolling(window=14).mean()
        loss = -delta.where(delta < 0, 0).rolling(window=14).mean()
        rs = gain / loss
        data['RSI'] = 100 - (100 / (1 + rs))
        data['BB_Middle'] = data['Close'].rolling(window=20).mean()
        data['BB_Std'] = data['Close'].rolling(window=20).std()
        data['BB_Upper'] = data['BB_Middle'] + 2 * data['BB_Std']
        data['BB_Lower'] = data['BB_Middle'] - 2 * data['BB_Std']
        high_low = data['High'] - data['Low']
        high_close = np.abs(data['High'] - data['Close'].shift())
        low_close = np.abs(data['Low'] - data['Close'].shift())
        tr = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)
        tr_smooth = tr.rolling(window=14).mean()
        data['Plus_DM'] = np.where((data['High'] - data['High'].shift(1)) > (data['Low'].shift(1) - data['Low']),
                                   (data['High'] - data['High'].shift(1)), 0)
        data['Minus_DM'] = np.where((data['Low'].shift(1) - data['Low']) > (data['High'] - data['High'].shift(1)),
                                    (data['Low'].shift(1) - data['Low']), 0)
        data['Plus_DI'] = 100 * (data['Plus_DM'].rolling(window=14).mean() / tr_smooth)
        data['Minus_DI'] = 100 * (data['Minus_DM'].rolling(window=14).mean() / tr_smooth)
        dx = 100 * np.abs(data['Plus_DI'] - data['Minus_DI']) / (data['Plus_DI'] + data['Minus_DI'])
        data['ADX'] = dx.ewm(span=14, adjust=False).mean()

        # Define signals
        data['MACD_Bullish'] = (data['MACD_Histogram'].shift(1) < 0) & (data['MACD_Histogram'] > 0)
        data['Stochastic_Bullish'] = (data['Stochastic_K'].shift(1) < data['Stochastic_D'].shift(1)) & \
                                     (data['Stochastic_K'] > data['Stochastic_D']) & (data['Stochastic_K'] < 30)
        data['RSI_Rebound'] = (data['RSI'].shift(1) < 30) & (data['RSI'] > 30)
        data['BB_Breakout'] = (data['Close'] > data['BB_Upper']) | (data['Close'] < data['BB_Lower'])
        data['Strong_Trend'] = data['ADX'] > 25

        # Backtest: Buy on signal, sell after 10 days
        hold_period = 10
        results = {}
        for signal in ['MACD_Bullish', 'Stochastic_Bullish', 'RSI_Rebound', 'BB_Breakout', 'Strong_Trend']:
            signals = data[data[signal]].index
            returns = []
            for signal_date in signals:
                if signal_date + pd.Timedelta(days=hold_period) in data.index:
                    buy_price = data.loc[signal_date, 'Close']
                    sell_price = data.loc[signal_date + pd.Timedelta(days=hold_period), 'Close']
                    trade_return = (sell_price - buy_price) / buy_price * 100
                    returns.append(trade_return)
            if returns:
                avg_return = np.mean(returns)
                success_rate = len([r for r in returns if r > 0]) / len(returns) * 100
                results[signal] = {'Avg Return': avg_return, 'Success Rate': success_rate}
            else:
                results[signal] = {'Avg Return': 0.0, 'Success Rate': 0.0}

        # Rank indicators by average return
        ranked_indicators = sorted(results.items(), key=lambda x: x[1]['Avg Return'], reverse=True)
        best_indicators = ranked_indicators[:3]  # Top 3 indicators
        return best_indicators
    except Exception as e:
        return None

# Plotting Functions
def plot_hourly_chart(ticker):
    spy = yf.Ticker(ticker)
    end_date = datetime.now(pytz.UTC)
    start_date_hourly = end_date - timedelta(days=7)
    df_hourly = spy.history(interval="1h", start=start_date_hourly, end=end_date)

    if not isinstance(df_hourly.index, pd.DatetimeIndex):
        df_hourly.index = pd.to_datetime(df_hourly.index)
    if df_hourly.index.tz is None:
        df_hourly.index = df_hourly.index.tz_localize('UTC')

    df_hourly['20MA'] = df_hourly['Close'].rolling(window=20).mean()
    df_hourly['50MA'] = df_hourly['Close'].rolling(window=50).mean()

    start_date_daily = end_date - timedelta(days=90)
    df_daily = spy.history(interval="1d", start=start_date_daily, end=end_date)
    if not isinstance(df_daily.index, pd.DatetimeIndex):
        df_daily.index = pd.to_datetime(df_daily.index)
    if df_daily.index.tz is None:
        df_daily.index = df_daily.index.tz_localize('UTC')

    recent_high = df_daily['High'].rolling(window=60).max().iloc[-1]
    recent_low = df_daily['Low'].rolling(window=60).min().iloc[-1]

    latest_close = df_hourly['Close'].iloc[-1]
    latest_20ma = df_hourly['20MA'].iloc[-1]
    latest_50ma = df_hourly['50MA'].iloc[-1]
    pct_diff_20ma = ((latest_close - latest_20ma) / latest_20ma) * 100
    pct_diff_50ma = ((latest_close - latest_50ma) / latest_50ma) * 100

    fig = plt.figure(figsize=(12, 7))
    ax = fig.add_subplot(111)

    if not df_hourly.empty:
        plot_high = df_hourly['High'].max()
        plot_low = df_hourly['Low'].min()
        show_support = recent_low >= plot_low and recent_low <= plot_high
        show_resistance = recent_high >= plot_low and recent_high <= plot_high

        apds = [
            mpf.make_addplot(df_hourly['20MA'], color='blue', label=f'20-hour MA ({pct_diff_20ma:.2f}%)', ax=ax),
            mpf.make_addplot(df_hourly['50MA'], color='orange', label=f'50-hour MA ({pct_diff_50ma:.2f}%)', ax=ax)
        ]

        mpf.plot(df_hourly, type='candle', ax=ax, volume=False, style='charles', addplot=apds)

        if show_support:
            ax.axhline(y=recent_low, linestyle='dashed', color='green', label='Support Zone')
        if show_resistance:
            ax.axhline(y=recent_high, linestyle='dashed', color='red', label='Resistance Zone')

        ax.set_title(f'{ticker_names[ticker]} ({ticker}) Hourly Chart (Last 7 Days)', pad=15)
        ax.set_xlabel('Date', labelpad=10)
        ax.set_ylabel('Price', labelpad=10)
        ax.legend()
        plt.setp(ax.get_xticklabels(), rotation=45, ha='right')
    else:
        ax.text(0.5, 0.5, "Hourly Data Unavailable", ha='center', va='center')
        ax.set_title(f'{ticker_names[ticker]} ({ticker}) Hourly Chart (Last 7 Days)', pad=15)
        ax.axis('off')

    plt.tight_layout(pad=2.0)
    plt.show()

def plot_weekly_chart(ticker):
    spy = yf.Ticker(ticker)
    end_date = datetime.now(pytz.UTC)
    start_date_long = end_date - timedelta(days=365 * 10)
    df_long = spy.history(interval="1d", start=start_date_long, end=end_date)

    df_weekly = df_long.resample('W').agg({
        'Open': 'first', 'High': 'max', 'Low': 'min', 'Close': 'last', 'Volume': 'sum'
    }).dropna()

    if not isinstance(df_weekly.index, pd.DatetimeIndex):
        df_weekly.index = pd.to_datetime(df_weekly.index)
    if df_weekly.index.tz is None:
        df_weekly.index = df_weekly.index.tz_localize('UTC')

    df_weekly['20MA'] = df_weekly['Close'].rolling(window=20).mean()
    df_weekly['50MA'] = df_weekly['Close'].rolling(window=50).mean()

    start_date_daily = end_date - timedelta(days=90)
    df_daily = spy.history(interval="1d", start=start_date_daily, end=end_date)
    if not isinstance(df_daily.index, pd.DatetimeIndex):
        df_daily.index = pd.to_datetime(df_daily.index)
    if df_daily.index.tz is None:
        df_daily.index = df_daily.index.tz_localize('UTC')

    recent_high = df_daily['High'].rolling(window=60).max().iloc[-1]
    recent_low = df_daily['Low'].rolling(window=60).min().iloc[-1]

    latest_close = df_weekly['Close'].iloc[-1]
    latest_20ma = df_weekly['20MA'].iloc[-1]
    latest_50ma = df_weekly['50MA'].iloc[-1]
    pct_diff_20ma = ((latest_close - latest_20ma) / latest_20ma) * 100
    pct_diff_50ma = ((latest_close - latest_50ma) / latest_50ma) * 100

    fig = plt.figure(figsize=(12, 7))
    ax = fig.add_subplot(111)

    plot_high = df_weekly['High'].max()
    plot_low = df_weekly['Low'].min()
    show_support = recent_low >= plot_low and recent_low <= plot_high
    show_resistance = recent_high >= plot_low and recent_high <= plot_high

    apds = [
        mpf.make_addplot(df_weekly['20MA'], color='blue', label=f'20-week MA ({pct_diff_20ma:.2f}%)', ax=ax),
        mpf.make_addplot(df_weekly['50MA'], color='orange', label=f'50-week MA ({pct_diff_50ma:.2f}%)', ax=ax)
    ]

    mpf.plot(df_weekly, type='candle', ax=ax, volume=False, style='charles', addplot=apds)

    if show_support:
        ax.axhline(y=recent_low, linestyle='dashed', color='green', label='Support Zone')
    if show_resistance:
        ax.axhline(y=recent_high, linestyle='dashed', color='red', label='Resistance Zone')

    ax.set_title(f'{ticker_names[ticker]} ({ticker}) Weekly Chart (10 Years)', pad=15)
    ax.set_xlabel('Date', labelpad=10)
    ax.set_ylabel('Price', labelpad=10)
    ax.legend()
    plt.setp(ax.get_xticklabels(), rotation=45, ha='right')

    plt.tight_layout(pad=2.0)
    plt.show()

def plot_technical_analysis(ticker):
    spy = yf.Ticker(ticker)
    end_date = datetime.now(pytz.UTC)
    start_date_ma = end_date - timedelta(days=365)
    df_daily_full = spy.history(interval="1d", start=start_date_ma, end=end_date)

    start_date_daily = end_date - timedelta(days=90)
    df_daily = df_daily_full.loc[start_date_daily:end_date].copy()  # Use .copy() to avoid SettingWithCopyWarning
    df_daily['20MA'] = df_daily['Close'].rolling(window=20).mean()
    df_daily['50MA'] = df_daily['Close'].rolling(window=50).mean()
    df_daily['Avg_Volume'] = df_daily['Volume'].rolling(window=10).mean()
    df_daily['Pct_Change'] = df_daily['Close'].pct_change() * 100

    start_date_long = end_date - timedelta(days=365 * 10)
    df_long = spy.history(interval="1d", start=start_date_long, end=end_date)

    for df in [df_daily, df_daily_full, df_long]:
        if not isinstance(df.index, pd.DatetimeIndex):
            df.index = pd.to_datetime(df.index)
        if df.index.tz is None:
            df.index = df.index.tz_localize('UTC')

    recent_high = df_daily['High'].rolling(window=60).max().iloc[-1]
    recent_low = df_daily['Low'].rolling(window=60).min().iloc[-1]

    monthly_returns = df_long['Close'].resample('ME').ffill().pct_change()
    seasonality = monthly_returns.groupby(monthly_returns.index.month).mean() * 100

    strikes = np.linspace(0.4, 1.0, 20)
    expirations = np.linspace(0.0, 1.0, 20)
    X, Y = np.meshgrid(strikes, expirations)
    Z = np.sin(X * 5) * np.exp(-Y) * 0.5 + 0.5

    atm_strike_idx = np.where(np.isclose(strikes, 1.0, atol=0.05))[0][0]
    atm_vols = Z[:, atm_strike_idx]
    iv_call = atm_vols.mean() + 0.05
    iv_put = atm_vols.mean() - 0.02
    iv_diff = iv_call - iv_put
    trend = "Bullish Skew" if iv_diff > 0 else "Bearish Skew"

    latest_close = df_daily['Close'].iloc[-1]
    latest_20ma = df_daily['20MA'].iloc[-1]
    latest_50ma = df_daily['50MA'].iloc[-1]
    pct_diff_20ma = ((latest_close - latest_20ma) / latest_20ma) * 100
    pct_diff_50ma = ((latest_close - latest_50ma) / latest_50ma) * 100

    significant_moves = df_daily[(abs(df_daily['Pct_Change']) > 2) &
                                 (df_daily['Volume'] > 1.5 * df_daily['Avg_Volume'])].tail(3)

    fig = plt.figure(figsize=(18, 12))

    # Daily Candlestick Chart with Volume
    ax1 = plt.subplot2grid((6, 3), (0, 0), rowspan=4, colspan=2)
    ax2 = plt.subplot2grid((6, 3), (4, 0), rowspan=1, colspan=2)
    plot_high = df_daily['High'].max()
    plot_low = df_daily['Low'].min()
    show_support = recent_low >= plot_low and recent_low <= plot_high
    show_resistance = recent_high >= plot_low and recent_high <= plot_high

    apds = [
        mpf.make_addplot(df_daily['20MA'], color='blue', label=f'20-day MA ({pct_diff_20ma:.2f}%)', ax=ax1),
        mpf.make_addplot(df_daily['50MA'], color='orange', label=f'50-day MA ({pct_diff_50ma:.2f}%)', ax=ax1)
    ]

    mpf.plot(df_daily, type='candle', ax=ax1, volume=ax2, style='charles', addplot=apds)

    if show_support:
        ax1.axhline(y=recent_low, linestyle='dashed', color='green', label='Support Zone')
    if show_resistance:
        ax1.axhline(y=recent_high, linestyle='dashed', color='red', label='Resistance Zone')

    for date in significant_moves.index:
        price = df_daily.loc[date, 'Close']
        ax1.plot(date, price, 'ro', markersize=10, label='Significant Move' if date == significant_moves.index[0] else "")

    ax1.set_title(f'{ticker_names[ticker]} ({ticker}) Technical Analysis (Daily - 3 Months)', pad=15)
    ax1.set_xlabel('Date (Last 3 Months)', labelpad=10)
    ax1.set_ylabel('Price (Index Level)', labelpad=10)
    ax2.set_ylabel('Volume', labelpad=10)
    ax1.legend()
    plt.setp(ax1.get_xticklabels(), rotation=45, ha='right')

    # Seasonality Bar Chart
    ax3 = plt.subplot2grid((6, 3), (5, 0), rowspan=1, colspan=2)
    ax3.bar(seasonality.index, seasonality.values, color='brown', alpha=0.7)
    ax3.set_xticks(range(1, 13))
    ax3.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
                        rotation=45, ha='right')
    ax3.set_title('Monthly Seasonality Over 10 Years', pad=15)
    ax3.set_xlabel('Month of Year', labelpad=10)
    ax3.set_ylabel('Avg Monthly Return (%)', labelpad=10)
    ax3.grid()

    # IV Surface Plot
    ax4 = plt.subplot2grid((6, 3), (0, 2), rowspan=2, colspan=1, projection='3d')
    surf = ax4.plot_surface(X, Y, Z, cmap='RdYlBu', alpha=0.9, edgecolor='none')
    ax4.set_xlabel('Strike Price (Normalized)', labelpad=10)
    ax4.set_ylabel('Time to Expiration (Years)', labelpad=10)
    ax4.set_zlabel('Implied Volatility (%)', labelpad=10)
    ax4.set_title('Implied Volatility Surface', pad=15)
    cbar = fig.colorbar(surf, ax=ax4, shrink=0.5, aspect=5)
    cbar.set_label('Volatility', rotation=270, labelpad=15)
    atm_strike = strikes[atm_strike_idx]
    ax4.plot([atm_strike] * len(expirations), expirations, atm_vols, color='black', linewidth=2, label='ATM Strike')
    ax4.legend()

    # IV Difference Text
    ax5 = plt.subplot2grid((6, 3), (2, 2), rowspan=1, colspan=1)
    iv_text = (
        "ATM IV Difference:\n"
        "  Call IV: {:.2f}\n".format(iv_call) +
        "  Put IV:  {:.2f}\n".format(iv_put) +
        "  IV Diff: {:.2f}\n".format(iv_diff) +
        "  Trend:   {}".format(trend)
    )
    ax5.text(0.05, 0.5, iv_text, fontsize=10, fontweight='bold', family='monospace',
             verticalalignment='center', horizontalalignment='left')
    ax5.axis('off')

    # Technical Indicators Summary
    result = check_technical_indicators(ticker)
    backtest_results = backtest_technical_indicators(ticker)
    if result:
        top_indicator = backtest_results[0][0] if backtest_results else "N/A"
        if result['Stock State'].startswith("Yes"):
            outcome = f"OUTCOME: Bullish Reversal Confirmed - Focus on {top_indicator}"
        elif result['Stock State'].startswith("Warning - Very Strong"):
            outcome = f"OUTCOME: Strong Bullish Reversal Likely - Watch {top_indicator}"
        elif result['Stock State'].startswith("Warning - Strong"):
            outcome = f"OUTCOME: Moderate Bullish Reversal Potential - Monitor {top_indicator}"
        elif result['Stock State'].startswith("Warning"):
            outcome = f"OUTCOME: Early Bullish Signs - Track {top_indicator}"
        else:
            outcome = f"OUTCOME: Bearish Trend Persists - Caution with {top_indicator}"

        summary_text = (
            f"{outcome}\n"
            "----------------------------\n"
            "Close Price:      {:.2f}\n".format(result['Close']) +
            "200EMA:           {:.2f} (% Diff: {:.2f}%)\n".format(result['200EMA'], result['Pct_Diff_200EMA']) +
            "20MA:             {:.2f} (% Diff: {:.2f}%)\n".format(result['20MA'], result['Pct_Diff_20MA']) +
            "50MA:             {:.2f} (% Diff: {:.2f}%)\n".format(result['50MA'], result['Pct_Diff_50MA']) +
            "RSI:              {:.2f}\n".format(result['RSI']) +
            "OBV:              {:.0f}\n".format(result['OBV']) +
            "BB Upper:         {:.2f}\n".format(result['BB_Upper']) +
            "BB Lower:         {:.2f}\n".format(result['BB_Lower']) +
            "ATR:              {:.2f}\n".format(result['ATR']) +
            "RVI:              {:.2f}\n".format(result['RVI']) +
            "ADX:              {:.2f}\n".format(result['ADX']) +
            "Volume Spike:     {}\n".format(result['Volume Spike']) +
            "MACD Bullish:     {}\n".format(result['MACD Bullish']) +
            "Stochastic Bull:  {}\n".format(result['Stochastic Bullish']) +
            "RSI Rebound:      {}\n".format(result['RSI Rebound']) +
            "BB Breakout:      {}\n".format(result['BB Breakout']) +
            "Strong Trend:     {}\n".format(result['Strong Trend']) +
            "Stock State:      {}\n".format(result['Stock State']) +
            "\nBacktest (2 Yr):\n"
        )
        if backtest_results:
            for i, (indicator, metrics) in enumerate(backtest_results, 1):
                summary_text += (
                    f"Top {i}: {indicator:<17} Avg Ret: {metrics['Avg Return']:.2f}% (Success: {metrics['Success Rate']:.1f}%)\n"
                )
        else:
            summary_text += "Backtest Failed\n"

        summary_text += "\nSignificant Moves (Last 3):\n"
        if not significant_moves.empty:
            for date, row in significant_moves.iterrows():
                summary_text += f"{date.strftime('%Y-%m-%d')}: Price Chg: {row['Pct_Change']:.2f}%, Vol: {row['Volume']:.0f}\n"
        else:
            summary_text += "No significant moves detected\n"

        ax6 = plt.subplot2grid((6, 3), (3, 2), rowspan=3, colspan=1)
        ax6.text(0.05, 0.5, summary_text, fontsize=10, fontweight='bold', family='monospace',
                 verticalalignment='center', horizontalalignment='left')
        ax6.axis('off')

    plt.tight_layout(pad=3.0)
    plt.show()

# Unit Tests
class TestTechnicalAnalysis(unittest.TestCase):
    def test_check_technical_indicators(self):
        result = check_technical_indicators('^GSPC')
        self.assertIsNotNone(result)
        self.assertIn('Close', result)
        self.assertIn('RSI', result)
        self.assertIn('BB_Upper', result)
        self.assertIn('ADX', result)

    def test_backtest_technical_indicators(self):
        result = backtest_technical_indicators('^GSPC')
        self.assertIsNotNone(result)
        self.assertEqual(len(result), 3)  # Top 3 indicators
        self.assertIn('MACD_Bullish', [ind[0] for ind in result])

# Main Execution
if __name__ == "__main__":
    ticker =  'CRWD'
    plot_hourly_chart(ticker)
    plot_weekly_chart(ticker)
    plot_technical_analysis(ticker)
    unittest.main(argv=[''], exit=False)
```
